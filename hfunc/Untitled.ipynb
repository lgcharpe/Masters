```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
weight = np.array([106.8, 106.3, 106.9, 105.9, 105.3, 103.9, 104.5, 103.8, 104.5, 103, 101.9, 101.6, 101.6, 102.9, 101.9, 100.9, 100.6, 100.6, 101.8, 100.4, 100, 99.7])
x = np.array([0, 1, 3, 5, 6, 7, 8, 9, 11, 18, 20, 21, 22, 25, 27, 28, 29, 30, 33, 34, 36, 37])
```

```python
plt.plot(x, weight)
```

```python
def best_fit_slope(xs,ys):
    m = (((np.mean(xs)*np.mean(ys)) - np.mean(xs*ys)) /
         ((np.mean(xs)*np.mean(xs)) - np.mean(xs*xs)))
    return m

m = best_fit_slope(x,weight)
print(m)
```
